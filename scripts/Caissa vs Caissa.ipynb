```python
import chess
import chess.pgn

from neural_caissa.board.state import State
from neural_caissa.board.move import move_player, get_pgn_from_moves

```

```python
state = State()
state.board.reset()
state.set_valuator('BaselineValuator')
print(state.board)
```

```python
idx = 1
moves = []
won = ''
while True:
    print('*' * 25 + f' move [{idx}]')
    move_str = move_player(state)
    moves.append(move_str)
    print(f'>>> white move ({move_str})')
    print(state.board)
    if state.board.is_game_over():
        print('game over... White wins')
        won = '1-0'
        break

    move_str = move_player(state)
    moves.append(move_str)
    print(f'\n>>> black move ({move_str})')
    print(state.board)
    if state.board.is_game_over():
        print('game over...Black wins')
        won = '0-1'
        break

    idx += 1
```

```python
print(get_pgn_from_moves(moves, 'baseline-algorithm', 'baseline-algorithm', won))
```
